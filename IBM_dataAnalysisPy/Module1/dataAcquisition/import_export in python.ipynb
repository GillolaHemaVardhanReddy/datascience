```python
# import pamdas
import pandas as pd
```

```python
# define the file path
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python
# import using .read_csv(path,header=None) # as we dont have header make sure to keep header=None
df = pd.read_csv(url,header=None)
```

```python
df.columns = headers
df.head()
```
