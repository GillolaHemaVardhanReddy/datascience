# *args unlimited positional arguments

creating functions that can take any number of arguments

```python
def add(a,b):
    print(a+b)
```

```python
add(3,4)
```

it is ok to give 2 arguments to add but what if we give 4 or 5 or n number of arguments then it raises error

#### solution:

```python
def add(*args):
    sum = 0
    for n in args:
        sum+=n
    print(sum)
```

```python
add(2,3,4)
```

```python
add(2,3,4,5)
```

hence solved yoo

here in above solution args is a tuple  

this tuple is generated by astric "*" any number of arguments are grouped into tuple

# **kargs unlimited keyword arguments

we can literally give any number of keyword arguments to the function and still it accepts them

**eg.**

```python
def calculate(**kargs):
    print(kargs,"type is : ",type(kargs))
```

```python
calculate(add=3,mul=5,sub=10,hema=20)
```

in above calculate we didn't specify those keyword arguments yet it accepted them 

because of  **kwargs 

this kwargs is basically turned into dictionary using **   
every input is grouped and creates a dictionary

```python
def calculate(n=0,**kargs):
    n += kargs["add"]
    n *= kargs["mul"]
    n -= kargs["sub"]
    print(n)
```

```python
calculate(add=3,mul=5,sub=10)
```

```python
calculate(10,add=3,mul=5,sub=10,hema=20)
```

#### better usage of kargs

```python
def calculate(n=0,**kargs):
    n += kargs.get("add")
    n *= kargs.get("mul")
    n -= kargs.get("sub")
    print(n)
```
