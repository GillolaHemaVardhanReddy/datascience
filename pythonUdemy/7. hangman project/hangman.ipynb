```python
import random as rd
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
```

```python
def hangMan():
    word_list = ["aardvark", "baboon", "camel"]
    word = rd.choice(word_list)
    guess = len(word)*'_' .split()
    loss = 0
    while(True):
        uguess = input(f"guess the letter to form {guess} : ").lower()
        found = False
        for i in range(len(word)):
            if word[i]==uguess:
                print("right guess")
                guess[i] = uguess
                found = True
        if not found :
            print("wrong guess")
            loss-=1
            print(stages[loss])
            if(loss<=-7):
                print("you lost")
                break
        elif '_' not in guess:
            print("you Won")
            break
        print(guess)
hangMan()
```
